## Introduction

Earlier this past January of 2018, Professor Laurie Santos led a spring semester class entitled *Psychology and the Good Life* for Yale University students, marking the course as the school's most popularly attended class of all-time enrolling close to 25% of the student population. Dr. Santos' class aimed to feature lectures on what things would or would not make people happy, along with contributors that might bring life-lasting satisfaction. From national media coverage by the New York Times to trending winds on social media, I begged the curious question - why are we crazed all of a sudden to find what happiness is? Is that intrinsic within us or is there an actual methodology to this "happiness"? And if so, what is it and does a happiness-driven life correlate to one's "success"?

In curiosity, I delved into **The World Happiness Report**, a United Nations programme started fairly recently in 2012, aiming to answer this very same question but on a global-level. They would gather these insights into the trajectory of bringing countries self-awareness of their own people's thoughts and future outlooks, ultimtaely as an indication of what a country can boast or improve upon. To do so, economists developed an in-depth survey with questions asking respective citizens what led them to the "happiness score" they had and the survey concluded their score into the sum of six distinct categories -- Economy (GDP), Family, Health and Life Expectancy, Freedom, Trust in the Government, and Generosity. 

Using Thinkful's Data Science prep course concepts, I analyzed my findings using various descriptive statistics and data visualizations.

## Dataset Overview

The dataset I analyzed ranged between the years of 2015-2017. Within the dataset, the Happiness Score is calculated by the sum of the six categorical variables the survey ascertained, along with a variable called "Dystopian_Residual". I did not focus all too much on this variable as it is defined by the World Happiness Report (WHR) simply as a "benchmark". The benchmark essentially represents the country "Dystopia" (as opposed to "Utopia") where happiness is virtually at the world's lowest, so it only serves as a baseline standard and holds no statistical value to one's survey or score. Thus, I created a column called "HappinessScore_woDystopia" to remove that "Dystopian_Residual" influence. However, for the sake of my dataset and the rest of my visualizations below, I chose to instead use the data's given "Happiness Score" as my tracker for Happiness Score because the scores will be more easily interpreted since its initial interpretation was based between 1-10 (10 being perfectly happy, 1 being absolutely miserable). 

While these six variables are intrinsically self-explanatory, it is important to realize that the denominator behind these variables is to frame your perception based on the citizen's outlook of that category. For example, the surveyor is not just interpreting their sense of "Economy_GDP" as the country's statistic but more along the lines of their perception of their country's economy, both on a local and national level and how that might dampen or improve in the future. Naturally, the aim of this analysis is to capture the relationship of these variables and not only how it contributes towards a particular Happiness Score but also the correlation between these variables themselves. Naturally, one's outlook on their own "Freedom" might be affected their "Trust in the Government" but it might not necessarily affect their own altruism in their "Generosity" score.

### *Conclusion*

Interestingly for all four regions, despite two of the three strongest correlated variables showing indications of improving Happiness Scores with rising in "Economy_GDP" and "Family" scores and worsening "Life Expectancy", Happiness Scores within the regions polarized in different directions. We also see in all four regions citizens having better "Trust in Government" and all four regions experiencing declining "Generosity" over the three year period, yet their happiness scores were different all around. 

This analysis did not show me much about the regions themselves. Happiness Scores for LATAM and the Carribean were also already much higher to begin with than the two improving regions so their scores were biased in some way to begin with. Quite possibly, however, this might lead to a different assumption about the entire dataset that many citizens globally might have similar outlooks as to what "happiness" is comprised of and that no matter what their own country experienced, the data might be intrinsically limited by nature in the evaluation of a citizen's happiness. It would be better to look less against the trends of data but instead more about the numbers themselves. For example, Central and Eastern Europe's "Economy GDP" numbers were higher than LATAM and Caribbean yet LATAM and Caribbean's "Freedom" scores were way higher - does this have any indication to a citizen's survey dependent on the culture they live in? What if a country's citizens have different priorities when they consider wealth versus freedom? We can also look at Sub-Saharan Africa and their life-expectancy scores and most likely infer that their health care system is far more limited and resource-deprived than in the other regions which might most likely have a large pull against their Happiness Score, regardless of how correlated that variable might be.

There are things the dataset can not explain and this question begs to probe at what factors within a region attributed to their happiness. The conclusion I got is that it is hard to confidently know without understanding what those surey questions were given how similarly these characteristics trended. There are more things that are systematic about the regions that affected these scores which is interesting to look more into in proposing for more research.

### *Conclusion*

For all the regions that experienced improving or worsening Happiness Scores, this last question helped sum up many of interesting assumptions about the dataset. Over the past two questions we talked holistically as to what could have been formulaic about a Happiness Score. But looking more deeply into the behavior of countries themselves, the variability of categories and what actually contributes to a citizen's Happiness Score is entirely dependent on where they live and the surrounding events that influence their day-to-day living. Most people react the same way in honing in on what makes them happy overall (top three correlated variables in "Economy_GDP", "Health_LifeExpectancy", and "Family"). But these variables undeniably change the more granularly you look into a country and their history. As every culture and country is unique so is their inpretation of happiness - so how can you really define happiness objectively?

## Proposal for More Research

My proposal for more research would include the lack of strength in altruism scores. "Generosity" for many countries had little effect on a country's Happiness Scores and it was more likely we saw altruism scores worsen - yet happiness was relatively unaffected. I would like to see more insight on the importance of "Generosity" and whether altruism has any affect on our well-being.

It is not a newfound revelation that giving to others and helping others in need, in principle, should be a thing that would elevate our own well-being. But ultimately there was no data to prove this notion in the dataset. For example, if there was data to track charitable dollars country by country, growth of non-profits (both on a national and global level), or measuring the total number of volunteer hours possibly by distinct people, there could maybe be better insight as to why this might be the case, whether that is an individual decision or a systematic flaw in programs not being readily available for people to engage in generous actions. This could further the possible assumption that a generous lifestyle may not elevate happiness posing interesting sociological questions. But as assumptions can only go so far, but my own thoughts like if issues like "Economy_GDP" negatively affect a country's people where the country is suffering from financial woes, it is more than likely for citizens to give less and ultimately, in the eyes of the citizen, justify this trend as less of a character defect but more so as a generally tough financial reason to do so.
